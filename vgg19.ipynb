```python
#Import Numpy, Keras Image and xception libraries
import numpy as np
from keras.preprocessing import image
from keras.applications import resnet
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
model = resnet.ResNet50()
model
```

```python
img = image.load_img('mountain.jpg', target_size=(224,224))
img
```

```python
#convert image to array
input_img = image.img_to_array(img)
input_img = np.expand_dims(input_img, axis=0)
input_img
```

```python
#pre_inputs = xception.preprocess_input(input_img)
predict_img = model.predict(input_img)
predict_img
```

```python
top_five_predict = resnet.decode_predictions(predict_img, top=5)
top_five_predict
```
